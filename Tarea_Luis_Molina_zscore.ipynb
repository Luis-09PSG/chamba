```python
import pandas as pd
import yfinance as yf
```

```python
sp500=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
sp500
```

```python
sp500IT=sp500[sp500['GICS Sector']=='Information Technology']
sp500IT
```

```python
tickers=sp500IT['Symbol'].to_list()
tickers
```

```python
df={'Ticker':[],'Working Capital':[],'Total Assets':[],'Retained Earnings':[],'EBIT':[],'Market Cap':[],'Current Price':[],'Long Term Liabilities':[],'Total Revenue':[]}
```

```python
for i in tickers:
  egyp=yf.Ticker(i).income_stmt
  bg=yf.Ticker(i).balance_sheet
  info=yf.Ticker(i).info
  wk=bg.loc['Working Capital'][0]
  ta=bg.loc['Total Assets'][0]
  re=bg.loc['Retained Earnings'][0]
  ebit=egyp.loc['EBIT'][0]
  mktcap=info.get('marketCap')
  currentprice=info.get('currentPrice')
  debt=bg.loc['Long Term Debt And Capital Lease Obligation'][0]
  tr=egyp.loc['Total Revenue'][0]
  df['Ticker'].append(i)
  df['Working Capital'].append(wk)
  df['Total Assets'].append(ta)
  df['Retained Earnings'].append(re)
  df['EBIT'].append(ebit)
  df['Market Cap'].append(mktcap)
  df['Current Price'].append(currentprice)
  df['Long Term Liabilities'].append(debt)
  df['Total Revenue'].append(tr)
data=pd.DataFrame(df)
data
```

```python
data.to_excel('sp500IT.xlsx')
from google.colab import files
files.download('sp500IT.xlsx')
```
